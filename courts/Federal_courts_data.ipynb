How did you find what the offense codes mean? I found a lookup table
in the documentation for a different (though similar) dataset of
federal courts data at the NACJD archive
You can find all the offense codes here in the codebook: http://www.icpsr.umich.edu/cgi-bin/file?comp=none&study=35491&ds=1&file_id=1171724&path=NACJD


```python
#So now tell me the district 
#that has the most convictions for interstate civil rights

#first I create a dataframe that has the civil rights code '9901'
#in any of the final offense columns using the | or "or operator"
civ_rights = cases[(cases.D2TOFFCD1 == '9901')|(cases.D2TOFFCD2 == '9901')|(cases.D2TOFFCD3 == '9901')|(cases.D2TOFFCD4 == '9901')|(cases.D2TOFFCD5 == '9901')]
```

```python
#now I'm checking if the DISP1 column has a value of 4,5,8 or 9,
#where the 1st offense is a '9901'.
#then, you'd add additional conditionals to catch where there were
#convictions for civil rights cases

#create a list of the codes for convictions used in the DISP1
#through DISP5 columns
convictions = [4,5,8,9]

#this is going to get messy. rather than putting these all in the filter
#i'm creating variables to store the filters in
#it'll make the last step easier
filter_1 = ((civ_rights.D2TOFFCD1 == '9901') & (civ_rights.DISP1.isin(convictions)))
filter_2 = ((civ_rights.D2TOFFCD2 == '9901') & (civ_rights.DISP2.isin(convictions)))
filter_3 = ((civ_rights.D2TOFFCD3 == '9901') & (civ_rights.DISP3.isin(convictions)))
filter_4 = ((civ_rights.D2TOFFCD4 == '9901') & (civ_rights.DISP4.isin(convictions)))
filter_5 = ((civ_rights.D2TOFFCD5 == '9901') & (civ_rights.DISP5.isin(convictions)))

#filter my civ_rights data frame with the five above ('1 or 2 or 3 or 4 or 5')
#then count the number of each value in the DISTRICT column
#thankfully value_counts automatically has a descending sort
civ_rights[filter_1|filter_2|filter_3|filter_4|filter_5].DISTRICT.value_counts()
```
